```python
lower_bound_vessel_1=0
upper_bound_vessel_1=0.04964
lower_bound_t = t.min()
upper_bound_t = t.max()
N_f=5000
N_u=len(t)

xi = np.linspace(lower_bound_vessel_1,upper_bound_vessel_1,N_u)[:,None]
ti = [[0.0]]*len(t)
xb = [[0.0]]*len(t)
tb = t
xc = lower_bound_vessel_1 + (upper_bound_vessel_1-lower_bound_vessel_1)*np.random.random((N_f))[:,None]
tc = lower_bound_t + (upper_bound_t-lower_bound_t)*np.random.random((N_f))[:,None]
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,12)
```

```python
t.min()
```

```python
plt.scatter(xc, tc, c ="blue",marker='x')
plt.scatter(xi,ti, c='green',marker='.')
plt.scatter(xb,tb, c='red',marker='.')
plt.xlabel("x",fontsize=25)
plt.ylabel("t",fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
